## 1. Variáveis
---

### O que é uma variável?

Uma variável é um **espaço na memória** destinado a guardar um valor que pode ser alterado durante a execução do programa.

> Python possui **tipagem dinâmica**: você não declara o tipo da variável — ele é inferido automaticamente em tempo de execução (diferente de C ou Java, por exemplo).

No exemplo abaixo, `x` recebe primeiro um número inteiro e depois uma string — e isso é totalmente válido em Python:

````python
x = 10
x = "miku"
# Esse código é válido graças à tipagem dinâmica do Python
````

- `x` é apenas um **rótulo** (referência), e não carrega um tipo em si
- os **valores** (`10` e `"miku"`) são os objetos que possuem tipo

## 2. Tipos de dados
---

### Visão geral dos tipos

| Categoria | Tipo | Descrição | Exemplo |
|---|---|---|---|
| Numérico | `int` | Números inteiros | `5`, `-10`, `100` |
| Numérico | `float` | Números decimais | `5.5`, `-3.14` |
| Numérico | `complex` | Números complexos | `2 + 3j` |
| Sequência | `str` | Texto (imutável) | `"abc"` |
| Sequência | `list` | Lista (mutável) | `[1, 2, 3]` |
| Sequência | `tuple` | Tupla (imutável) | `(1, 2, 3)` |
| Mapeamento | `dict` | Dicionário chave→valor | `{"nome": "Miku"}` |
| Lógico | `bool` | Valores lógicos | `True`, `False` |

````python
ano = 2007
nome = "Hatsune Miku"
profissao = "cantora virtual"
genero_musical = "vocaloid"

print( "A", nome, "é uma", profissao, "do gênero", genero_musical, "que surgiu no ano de ", ano)
print(type(ano)) # o comando type() é usado para verificar o tipo de dado de uma variável
print(type(nome))
````

### Saída de dados

Existem quatro formas principais de exibir variáveis com `print()`:

| Forma | Sintaxe | Observação |
|---|---|---|
| Vírgulas | `print("Olá", nome)` | Simples, adiciona espaço automático |
| `format()` | `"Olá {}".format(nome)` | Mais flexível |
| `%` | `"Olá %s" % nome` | Estilo antigo (C-like) |
| **f-string** | `f"Olá {nome}"` | Mais moderna e legível |

````python
# Forma 1 - concatenação com vírgula (mais simples, a que utilizamos até agora)
print("A", nome, "é uma", profissao, "do gênero", genero_musical, "que surgiu no ano de", ano)


# Forma 2 - usando format()
print("A {} é uma {} do gênero {} que surgiu no ano de {}.".format(nome, profissao, genero_musical, ano))


# Forma 3 - usando operador %
print("A %s é uma %s do gênero %s que surgiu no ano de %d." % (nome, profissao, genero_musical, ano))


print("\n----- Com f-string -----")


# Forma 4 - f-string (mais moderna e recomendada)
print(f"A {nome} é uma {profissao} do gênero {genero_musical} que surgiu no ano de {ano}.")


# Expressões dentro da f-string
proximo_ano = ano + 1
print(f"No ano de {proximo_ano}, ela completou {proximo_ano - ano} ano(s) de carreira.")


# Formatação de números
nota = 9.87654
print(f"A nota é {nota}, e ela formatada é: {nota:.2f}")  # 2 casas decimais
````

## 3. Tipo Numérico
---

### Operações aritméticas

Os operadores disponíveis para tipos numéricos são:

| Operador | Operação | Exemplo | Resultado |
|---|---|---|---|
| `+` | Adição | `3 + 2` | `5` |
| `-` | Subtração | `3 - 2` | `1` |
| `*` | Multiplicação | `3 * 2` | `6` |
| `/` | Divisão (float) | `5 / 2` | `2.5` |
| `//` | Divisão inteira | `5 // 2` | `2` |
| `%` | Resto (módulo) | `5 % 2` | `1` |
| `**` | Potência | `2 ** 3` | `8` |

**1. Python é fortemente tipado**

Mesmo com tipagem dinâmica, operações entre tipos **incompatíveis geram erro**:

````python
x = "10"  # x é do tipo string
y = 5     # y é do tipo inteiro
soma = x + y  # ERRO! Pois não é possível somar uma string com um inteiro
````

**2. Atenção ao tipo do resultado**

Ao dividir dois `int`, o resultado é sempre um `float` — mesmo que a divisão seja exata. Teste abaixo:

````python
x = 5
y = 2
divisao = x / y
print(divisao) # Valor da divisão, que é um número decimal (float) mesmo que x e y sejam inteiros (int)
````

````python
print(x // y)  # Operador de divisão inteira (floor division), que retorna o resultado da divisão sem a parte decimal
````

````python
print(x % y) # Operador de módulo, que retorna o resto da divisão entre x e y
````

````python
print(y ** x) # Operador de potência, que calcula y elevado à potência de x
````

````python
from math import sqrt
print(sqrt(16)) # Função sqrt servepara calcular a raiz quadrada de um número, que é importada da biblioteca math
````

### Conversão de tipos (`casting`)

É possível converter um valor para outro tipo usando as funções `int()`, `float()`, `str()`, `bool()`, etc.

> `int()` **trunca** (não arredonda) a parte decimal: `int(5.9)` → `5`

````python
print(int("10"))     # A string "10" é convertida para o número inteiro 10
print(int(5.9))      # 5  (trunca a parte decimal)
print(int(True))     # True é convertido para 1
print(int(False))    # False é convertido para 0
````

## 4. Tipo Booleano (`bool`)
---

O tipo `bool` representa valores de **verdade**: `True` ou `False`.

Booleanos são o resultado de comparações e expressões lógicas, e serão muito usados nas condicionais (próxima apostila!).

Os **operadores de comparação** produzem booleanos:

| Operador | Significado | Exemplo |
|---|---|---|
| `==` | igual a | `x == 10` |
| `!=` | diferente de | `x != 10` |
| `>` | maior que | `x > 5` |
| `<` | menor que | `x < 5` |
| `>=` | maior ou igual | `x >= 5` |
| `<=` | menor ou igual | `x <= 5` |

E os **operadores lógicos** combinam booleanos:

| Operador | Resultado |
|---|---|
| `and` | `True` se **ambos** forem verdadeiros |
| `or` | `True` se **pelo menos um** for verdadeiro |
| `not` | **Inverte** o valor lógico |

````python
print(f"----- Valores booleanos (verdadeiro ou falso )----- ")
possui_metadados = True
print(f"O documento possui metadados completos? \n Resposta: {possui_metadados} (tipo: {type(possui_metadados)})")
````

````python
print(f"----- Valores booleanos -- comparação (verdadeiro ou falso) ----- ")
print(f"O número de documentos é maior que 1000? Resposta: {numero_documentos > 1000}\n")

bool1 = ano_publicacao == 1900
# O operador de comparação == é usado para verificar se o valor de ano_publicacao é igual a 1900. O resultado dessa comparação será um valor booleano (True ou False)
# A resposta dessa comparação é False, pois o valor de ano_publicacao é 1888, que não é igual a 1900. Portanto, bool1 armazena o valor False

print(f"O ano de publicação é igual a 1900? Resposta: {bool1}")
print(f"A porcentagem digitalizada é menor ou igual a 90%? Resposta:{porcentagem_digitalizada <= 90.0}")
````

````python
print(f"----- Valores booleanos (operadores lógicos AND OR e NOT )----- ")

print("\n--- Operador AND ---")
# AND: ambas as condições precisam ser verdadeiras
print(numero_documentos > 1000 and numero_artigos > 50)
# O resultado dessa expressão é False, pois a primeira condição (numero_documentos > 1000) é verdadeira
# Mas a segunda condição (numero_artigos > 50) é falsa. Para que o operador AND retorne True, ambas as condições precisam ser verdadeiras


print("\n--- Operador OR ---")
# OR: pelo menos uma condição precisa ser verdadeira
print(numero_documentos > 1000 or numero_artigos > 50)
# O resultado dessa expressão é True, pois a primeira condição (numero_documentos > 1000) é verdadeira


print("\n--- Operador NOT ---")
# NOT: inverte o valor lógico da expressão
print(not (ano_publicacao > 2000))
# O resultado dessa expressão é True, pois a expressão dentro do parênteses (ano_publicacao > 2000) é falsa, e o operador NOT inverte esse valor para True
````

## 5. Strings (`str`)
---

**Strings** são sequências de caracteres delimitadas por aspas simples `'texto'` ou duplas `"texto"`.

As principais operações com strings são:

| Operação | Sintaxe | Exemplo | Resultado |
|---|---|---|---|
| Concatenar | `a + b` | `"Hatsune" + " Miku"` | `"Hatsune Miku"` |
| Repetir | `a * n` | `"ha" * 3` | `"hahaha"` |
| Indexar | `s[i]` | `"Miku"[0]` | `"M"` |
| Fatiar | `s[i:j]` | `"Miku"[0:2]` | `"Mi"` |
| Tamanho | `len(s)` | `len("Miku")` | `4` |


> Índices negativos contam a partir do final: `s[-1]` é o último caractere.

````python
print("\n--- Exemplo de Manipulação de Strings (Hatsune Miku) ---\n")

nome = "Hatsune Miku"
profissao = "cantora virtual"

# Concatenação
frase = nome + " é uma " + profissao

# Repetição
repetir = nome * 3

# Indexação (acessando caracteres individuais)
primeira_letra = nome[0]
ultima_letra = nome[-1] # Por que -1? Porque a contagem foi feita de trás para frente, ou seja, o último caractere tem índice -1, o penúltimo tem índice -2, e assim por diante.

# Fatiamento (slicing)
primeiras_letras = nome[0:7]  # O fatiamento é feito usando a sintaxe [início:fim], onde o índice de início é inclusivo e o índice de fim é exclusivo. Então, nome[0:7] retorna os caracteres do índice 0 ao 6
ultimas_letras = nome[-4:]  # O fatiamento também pode ser feito usando índices negativos. nome[-4:] retorna os últimos 4 caracteres da string, ou seja, do índice -4 até o final da string.

# Exibindo os resultados
print("Concatenação:", frase)
print("Repetição:", repetir)
print("Primeira letra:", primeira_letra)
print("Última letra:", ultima_letra)
print("Primeiras letras:", primeiras_letras)
print("Últimas letras:", ultimas_letras)
````

````python
print("----- Mais exemplos de manipulação de string -----\n")

texto = "Apostila digital de Python para iniciantes"

print(texto.upper())

print(texto.lower())

print(texto.capitalize())

print(texto.replace("digital", "física")) # Sintaxe: string.replace("texto a ser substituído", "novo texto")

print(texto.split())
#  O método split() divide a string em uma lista de palavras, usando o espaço como separador por padrão
# Para mudar o separador, basta passar o caractere desejado como argumento, por exemplo: texto.split("a") dividiria a string em partes usando a letra "a" como o separador

print(texto, "possui", len(texto), "caracteres totais")
````

Também podemos acessar **caracteres individuais** por índice e medir o **tamanho** da string com `len()`:

````python
print("----- Mais outros exemplos de manipulação de string -----\n")

print(texto[0])
print(texto[1])
print(texto[2]), print(texto[3]), print(texto[4]), print(texto[5]), print(texto[6]), print(texto[7]), print(texto[8])
print (len(texto))
````

### Principais métodos de strings

Strings possuem vários métodos embutidos. Os mais usados:

| Método | O que faz | Exemplo | Resultado |
|---|---|---|---|
| `.lower()` | Minúsculas | `"ABC".lower()` | `"abc"` |
| `.upper()` | Maiúsculas | `"abc".upper()` | `"ABC"` |
| `.capitalize()` | Primeira letra maiúscula | `"python".capitalize()` | `"Python"` |
| `.strip()` | Remove espaços nas bordas | `"  oi  ".strip()` | `"oi"` |
| `.replace(a, b)` | Substitui ocorrências | `"banana".replace("a","o")` | `"bonono"` |
| `.split(sep)` | Divide em lista | `"a,b,c".split(",")` | `["a","b","c"]` |
| `.join(lista)` | Une lista em string | `"-".join(["a","b"])` | `"a-b"` |
| `.find(sub)` | Posição da substring | `"python".find("th")` | `2` |
| `.startswith(s)` | Começa com? | `"python".startswith("py")` | `True` |
| `.endswith(s)` | Termina com? | `"python".endswith("on")` | `True` |
