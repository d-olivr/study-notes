# 2. Condicionais

## Conceito
---

Estruturas condicionais são utilizadas quando vê-se a necessidade de adicionar condições no código, por exemplo:

- **Se** está chovendo, abro guarda-chuva.
- **Se** a nota for maior que 5, o aluno passou; **senão**, reprovou.

## Comando `if`
---

A estrutura mais básica é o `if` ("se", em inglês)

```python
temperatura = 35

# Sintaxe
if temperatura > 30:
    print("Está calor!")
```

Ao mudar o valor de `temperatura` para `25`, o que aconteceria? 
Nada, já que a condição é **falsa** e o bloco não é executado.

## Comandos `if` + `else`
---

É possivel executar código **nos dois casos**, quando a condição é verdadeira *e* quando é falsa

```python
nota = 6.5

if nota >= 5:
    print("Aprovado")
else:
    print("Reprovado")
```

## Comandos `if` + `elif` + `else`
---

Às vezes temos **mais de duas possibilidades**. O `elif` (abreviação de "else if", ou "senão se") permite encadear várias condições

> O Python testa as condições de cima pra baixo e executa apenas o primeiro bloco verdadeiro. Os demais são ignorados.

```python
print("----- Teste de Notas -----")
nota = 2.2

if nota >= 9:
    conceito = "A"
elif nota >= 7:
    conceito = "B"
elif nota >= 5:
    conceito = "C"
else:
    conceito = "D"

print(f"Nota: {nota}, Conceito: {conceito}")
```

## Operadores de comparação
---

É possível juntar condicionais com operadores de comparação (expressões que retornam `True` ou `False`)

| Operador | Significado        | Exemplo       |
|----------|--------------------|---------------|
| `==`     | igual a            | `x == 10`     |
| `!=`     | diferente de       | `x != 10`     |
| `>`      | maior que          | `x > 5`       |
| `<`      | menor que          | `x < 5`       |
| `>=`     | maior ou igual     | `x >= 5`      |
| `<=`     | menor ou igual     | `x <= 5`      |

> Não confundir `=` (atribuição) com `==` (comparação)

```python
print("----- Operadores de Comparação -----")
x = 10

print(x == 10)   # True
print(x != 10)   # False
print(x > 7)     # True
print(x <= 9)    # False
```

## Operadores lógicos
---

Também é possível combinar condições com os operadores lógicos:

| Operador | Significado                              |
|----------|------------------------------------------|
| `and`    | verdadeiro se ambas forem verdadeiras |
| `or`     | verdadeiro se pelo menos uma for      |
| `not`    | inverte o valor lógico                |

```python
idade = 22
tem_carteira = True

if idade >= 18 and tem_carteira:
    print("Pode dirigir")
else:
    print("Não pode dirigir")
```

```python
dia = "sabado"

if dia == "sabado" or dia == "domingo":
    print("Final de semana")
else:
    print("Dia útil")
```

```python
logado = False

if not logado:
    print("Por favor, faça login.")
```

## Aninhamento
---

É possível colocar um `if` dentro de outro. Isso se chama **aninhamento** (do inglês nesting)

> Use o aninhamento com moderação, estruturas muito aninhadas ficam difíceis de ler e debugar

```python
# Exemplo de uso de condicionais aninhados
saldo = 500
valor_compra = 200
conta_ativa = True

if conta_ativa:
    if saldo >= valor_compra:
        print("Compra aprovada")
    else:
        print("Saldo insuficiente")
else:
    print("Conta inativa. Entre em contato com o banco")
```

## Condicional em uma linha (expressão ternária)
---

Para casos simples, Python permite escrever um `if/else` em uma única linha

```python
idade = 1

# Sintaxe: variável = valor_se_verdadeiro if condição else valor_se_falso
status = "maior de idade" if idade >= 18 else "menor de idade"
print(status)
```

## Valores *truthy* e *falsy*
---

Em Python, qualquer valor pode ser usado como condição, não só `True` e `False`.

São considerados **falsos** (`falsy`):
- `False`
- `0` e `0.0`
- `""` (string vazia)
- `[]`, `{}`, `()` (coleções vazias)
- `None`

Todo o resto é considerado **verdadeiro** (`truthy`).

```python
nome = ""

if nome:
    print(f"Olá, {nome}!")
else:
    print("Nome não informado.")
```

```python
carrinho = ["maçã", "banana", "laranja", "pera"]

if carrinho:
    print(f"Seu carrinho tem {len(carrinho)} item(ns).")
else:
    print("Seu carrinho está vazio.")
```

## Operador `in`
---

O `in` verifica se um valor está dentro de uma sequência (string, lista, etc.). É muito útil em condicionais

```python
cesta_de_frutas = ["maçã", "banana", "uva"]

fruta_buscada = "banana"

if fruta_buscada in cesta_de_frutas:
    print(f"{fruta_buscada} está disponível na cesta.")
else:
    print(f"{fruta_buscada} não encontrada na cesta.")
```
